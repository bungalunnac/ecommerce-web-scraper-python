```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://books.toscrape.com/"
respon = requests.get(url)
soup = BeautifulSoup(respon.text, 'html.parser')

kumpulan_buku = soup.find_all('article', class_='product_pod')

list_data_buku = []

for buku in kumpulan_buku:
    judul = buku.h3.a['title']
    harga = buku.find('p', class_='price_color').text

    data_buku = {
        "Judul Buku": judul,
        "Harga": harga
    }

    list_data_buku.append(data_buku)

df = pd.DataFrame(list_data_buku)

df.to_csv("hasil_scraping_buku.csv", index=False)

print("Selesai! File 'hasil_scraping_buku.csv' berhasil dibuat di folder kamu!")
print("\nBerikut intip isi tabelnya:")
print(df.head())
```
